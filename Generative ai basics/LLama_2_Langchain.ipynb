```python
from langchain.llms import HuggingFacePipeline
from transformers import AutoTokenizer
import torch
import transformers
```

## LLama 2 loading

```python
# model = "meta-llama/Llama-2-7b-chat-hf" Because this model of meta needs first permission
model = "daryl149/llama-2-7b-chat-hf"
```

```python
# Importing the tokenizer
tokenizer = AutoTokenizer.from_pretrained(model)
```

```python
pipeline = transformers.pipeline(
    model = model,
    tokenizer = tokenizer,
    torch_dtype = torch.float16,
    device_map = "auto",
    max_length = 1024,
    do_sample = True,
    top_k = 10,
    num_return_sequences = 1,
    eos_token_id = tokenizer.eos_token_id
) ### Loading of my model
```

```python
llm = HuggingFacePipeline(pipeline = pipeline, model_kwargs = {'temperature':0.5})
```

```python
prompt = "What is the process to become a real good strategist ?"
```

```python
print(llm(prompt))
```

## Prompt template

```python
from langchain.prompts import PromptTemplate # This is the prompt template function
from langchain.chains import LLMChain # This is the chain function
```

```python
prompt_template = PromptTemplate(
    input_variables = ["Strategy"],
    template = "Give me the summary of {Strategy} ?"
)
```

```python
input_prompt = prompt_template.format(Strategy = "War Art Sun Tzu")
```

```python
print(input_prompt)
```

```python
chain = LLMChain(llm = llm, prompt = prompt_template, verbose = True)
response = chain.run(input_prompt)
print(response)
```
